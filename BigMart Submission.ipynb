```python
# Import sklearn.preprocessing.StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler, then apply it to the features
scaler = MinMaxScaler() # default=(0, 1)
numerical = ["Item_MRP", "Item_Outlet_Sales", "Item_Weight", "Outlet_Establishment_Year"]

features_log_minmax_transform = pd.DataFrame(big_mart)
features_log_minmax_transform[numerical] = scaler.fit_transform(big_mart[numerical])

# Show an example of a record with scaling applied
features_log_minmax_transform.head(n = 5)
```

```python
#Import library:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#New variable for outlet
big_mart['Outlet'] = le.fit_transform(big_mart['Outlet_Identifier'])
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type','Outlet']
le = LabelEncoder()
for i in var_mod:
    big_mart[i] = le.fit_transform(big_mart[i])
```

```python
XCols = ['Item_MRP', 'Item_Weight',
       'Outlet_Establishment_Year', 'Type',
       'Item_Identifier_Drinks', 'Item_Identifier_Food',
       'Item_Identifier_Non-Consumable', 'Item_Fat_Content_0',
       'Item_Fat_Content_1', 'Outlet_Location_Type_0',
       'Outlet_Location_Type_1', 'Outlet_Location_Type_2', 'Outlet_Size_0',
       'Outlet_Size_1', 'Outlet_Size_2', 'Outlet_Type_0', 'Outlet_Type_1',
       'Outlet_Type_2', 'Outlet_Type_3', 'Outlet_0', 'Outlet_1', 'Outlet_2',
       'Outlet_3', 'Outlet_4', 'Outlet_5', 'Outlet_6', 'Outlet_7', 'Outlet_8',
       'Outlet_9']

YCols = ["Item_Outlet_Sales"]
```

```python
XX = big_mart.loc[big_mart["Type"]=="train",XCols]
yy = big_mart.loc[big_mart["Type"]=="train",YCols]

# XX.drop(["Type"],axis=1,inplace=True)
# yy.drop(["Type"],axis=1,inplace=True)
print(XX.shape)
print(yy.shape)
XX.head()
```

```python
XX.drop(["Type"],axis=1,inplace=True)
# yy.drop(["Type"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
XX_train, XX_test, yy_train, yy_test = train_test_split(XX, yy, test_size=0.3, random_state=5)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression(fit_intercept=True,normalize=True)
reg.fit (XX_train,yy_train)
print(reg.intercept_)
reg.coef_
```

```python
yy_predicted = reg.predict(XX_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_true=yy_test, y_pred=yy_predicted)
```

```python
from sklearn.metrics import r2_score
r2_score(yy_test, yy_predicted)
```
